## 8. Word Count

Given a sentence, return a dictionary with word counts.

Input: "data engineering data pipeline"  
Output: {'data': 2, 'engineering': 1, 'pipeline': 1}


```python
input="data engineering data pipeline"
input_list=input.split(" ")
print(input_list)
count_dict={}
for i in input_list:
    if i not in count_dict:
        count_dict[i]=1
    else:
        prev_count=count_dict[i]
        count_dict[i]=prev_count+1
print(count_dict)
```
