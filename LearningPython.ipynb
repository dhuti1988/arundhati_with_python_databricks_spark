```python
def check_prime(x):
    is_prime = True
    for i in range(2, int(math.sqrt(x))):
        if x % i == 0:
            is_prime = False
            break
    return is_prime

check_prime(12)
```

```python
def duplicate_list(l1):
    l2=l1[0:1]
    for i in range(1,len(l1)):
        if l1[i]==l1[i-1]:
            i=i+1
        else:
            l2.append(l1[i])
    return l2

duplicate_list([1,2,2,3,4,4,5])
```

```python
import pandas as pd
import numpy as np

# -------------------------------
# A. Generate Customers Dataset
# -------------------------------
np.random.seed(42)

customer_ids = range(1001, 1051)
names = [f"Customer_{i}" for i in customer_ids]
ages = np.random.randint(18, 65, size=50)
cities = np.random.choice(["New York", "Boston", "Chicago", "Dallas", "Seattle"], size=50)

customers = pd.DataFrame({
    "customer_id": customer_ids,
    "name": names,
    "age": ages,
    "city": cities
})

customers.to_csv("customers.csv", index=False)


# -------------------------------
# B. Generate Products Dataset
# -------------------------------
product_ids = range(2001, 2011)
product_names = [f"Product_{i}" for i in product_ids]
categories = np.random.choice(["Electronics", "Clothing", "Grocery"], size=10)
prices = np.random.randint(5, 500, size=10)

products = pd.DataFrame({
    "product_id": product_ids,
    "product_name": product_names,
    "category": categories,
    "price": prices
})

products.to_csv("products.csv", index=False)


# -------------------------------
# C. Generate Orders Dataset
# -------------------------------
order_ids = range(3001, 3301)
order_customer_ids = np.random.choice(customer_ids, size=300)
order_product_ids = np.random.choice(product_ids, size=300)
quantities = np.random.randint(1, 5, size=300)
order_dates = pd.date_range("2024-01-01", periods=300).strftime("%Y-%m-%d")

orders = pd.DataFrame({
    "order_id": order_ids,
    "customer_id": order_customer_ids,
    "product_id": order_product_ids,
    "quantity": quantities,
    "order_date": order_dates
})

orders.to_csv("orders.csv", index=False)
```

## Learn Classes

```python
class BankAccount:
    def __init__(self, account_number, holder_name, balance=0):
        self.account_number = account_number
        self.holder_name = holder_name
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited ${amount}. New balance: ${self.balance}")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds.")

    def get_balance(self):
        return self.balance
```

```python
accj = BankAccount("ACC123", "Joyjit Chowdhury", 500)
```

```python
accj.deposit(200)
accj.balance = 0
```

```python
# Create an account object
acc = BankAccount("ACC123", "Joyjit Chowdhury", 500)

# Make a deposit
acc.deposit(200)  
# Output → Deposited $200. New balance: $700

# Make a withdrawal
acc.withdraw(150)
# Output → Withdrew $150. Remaining balance: $550

# Check balance
print(acc.get_balance())  
# Output → 550
```
