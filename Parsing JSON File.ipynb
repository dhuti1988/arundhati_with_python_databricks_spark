# Parsing JSON Files

```python
import json
import pandas as pd

file_path='/Workspace/Users/arundhatidaschowdhury@gmail.com/arundhati_with_python_databricks_spark/data/transaction.json'
transaction=pd.read_json(file_path)
display(transaction)
```

```python
# Normalize the 'customer' column to flatten nested customer data
normalise_transaction = pd.json_normalize(transaction["customer"])

# Concatenate the normalized customer data with the rest of the transaction data
expanded_transaction = pd.concat([transaction.drop(columns=["customer"]), normalise_transaction], axis=1)

# Explode the 'items' column to create a separate row for each item
expanded_transaction = expanded_transaction.explode("items").reset_index(drop=True)

# Normalize the 'items' column to flatten nested item data
expand_item = pd.json_normalize(expanded_transaction["items"]).reset_index(drop=True)

# Concatenate the expanded item data with the rest of the transaction data
complete_expand = pd.concat([expanded_transaction.drop(columns=["items"]), expand_item], axis=1)

# Display the fully expanded transaction DataFrame
display(complete_expand)
```
