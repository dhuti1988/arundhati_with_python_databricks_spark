```python
import pandas as pd

#Using functions in pandas
def get_initial_name(fname,lname):
    return fname[0] + "." + lname


data = {
    "fname": ["John", "Jane", "Alice", "Bob", "Eve"],
    "lname": ["Doe", "Smith", "Johnson", "Brown", "Davis"]
}

df_sample = pd.DataFrame(data)
#df_sample["full_name"]=df_sample["f"].str[0]+"."+df_sample["lname"]
display(df_sample)
df_sample["full_name"] = df_sample.apply(lambda row: get_initial_name(row['fname'], row['lname']), axis=1)

display(df_sample)
```

```python
import os
import json
import requests

file = "/Workspace/Users/arundhatidaschowdhury@gmail.com/data/SampleDataFile.json"

API_KEY = "77625a8592944180861ba4c6d9008328"
url = "https://api.geoapify.com/v1/geocode/autocomplete"

def call_api(addr_text):
    params = {
        "text": addr_text,
        "apiKey": API_KEY
        }
    response = requests.get(url, params=params)
    data = response.json()
    return data
```

- Create column addr_text
- Call API with this new column 
- Create another column address_enriched

```python
def enrich_address(address_json):
    # prepare string 
    addr_text = " ".join(address_json.values())
    # Call API to get new address
    api_response = call_api(addr_text)
    #Extract new address from API response 
    new_data=api_response["features"][0]["properties"]
    address_enriched = {
        "add1": new_data["address_line1"],
        "add2": "",
        "city": new_data["city"],
        "state": new_data["state_code"],
        "zip": new_data["postcode"] 
        }
    return(address_enriched)

df["address_enriched"] = df.apply(lambda row: enrich_address(row['address']), axis=1)    
display(df)
```
