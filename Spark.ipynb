```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("TEST_JC").getOrCreate()
```

```python
from pyspark.sql import Row

# Create DataFrame with fname, lname, and salary columns
data = [
    Row(fname="John", lname="Doe", salary=50000),
    Row(fname="Jane", lname="Smith", salary=60000)
]
df = spark.createDataFrame(data)
```

```python
display(df)
```

```python
from pyspark.sql.functions import col
```

```python
# spark expression
df = df.withColumn("salary", col("salary") + 100)
```

```python
display(df)
```

```python
# Basic transform: Add a new column 'col1_squared'
from pyspark.sql.functions import col
transformed_df = df.withColumn('col1_squared', col('col1') * col('col1'))
```

```python

# Write the transformed DataFrame to CSV
transformed_df.write.csv('/tmp/transformed_sample_1000.csv', header=True, mode='overwrite')
```
